```python
# Busco elementos duplicado
export.duplicated()
```

```python
# Mostrar filas duplicadas antes de eliminarlas
duplicates_before = export[export.duplicated(keep=False)]
print("Duplicados antes de eliminarlos:")
print(duplicates_before)
```

```python
# Eliminamos duplicados
export_cleaned = export.drop_duplicates()
print(export_cleaned)
```

```python

export_cleaned = pd.DataFrame(export_cleaned)

for col in export_cleaned.columns:
    if col != 'distributor':
        export_cleaned[col] = export_cleaned[col].replace({'\$': '', '\.': '', ',00': ''}, regex=True).str.replace('.', '').astype(float)

sum_values = export_cleaned.drop(columns=['distributor']).sum()

sum_values_millions = sum_values / 1_000_000

print(sum_values_millions.apply(lambda x: f"{x:.2f} millones"))
```

```python
sales = pd.read_csv("sales_paraguay.csv")
sales
```

```python
sales.duplicated()
```

```python
sales_cleaned = sales.drop_duplicates()
print(sales_cleaned)
```

```python
sales_cleaned = pd.DataFrame(sales_cleaned)

sales_cleaned = sales_cleaned.iloc[:-1]

def clean_currency(x):
    return float(x.replace('$', '').replace('.', '').replace(',', '.'))

for col in sales_cleaned.columns:
    if col != 'distributor':
        sales_cleaned[col] = sales_cleaned[col].apply(clean_currency)

sum_values = sales_cleaned.drop(columns=['distributor']).sum()

sum_values_millions = sum_values / 1_000_000

pd.set_option('display.float_format', lambda x: '%.2f' % x)
print(sum_values_millions.apply(lambda x: f"{x:.2f} millones"))
```

```python
result_df = pd.DataFrame({
    'total_exports': export_cleaned.iloc[0],
    'total_ventas': sales_cleaned.iloc[0]
})

result_df['ratio_ventas/exports'] = result_df['total_ventas'] / result_df['total_exports']

pd.set_option('display.float_format', lambda x: '%.2f' % x)

print(result_df)
```

```python
import matplotlib.pyplot as plt

materiales = result_df.index
exportaciones = result_df['total_exports']
ventas = result_df['total_ventas']

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.4
opacity = 0.8
index = range(len(materiales))
rects1 = plt.bar(index, exportaciones, bar_width, alpha=opacity, color='b', label='Exportaciones')
rects2 = plt.bar([i + bar_width for i in index], ventas, bar_width, alpha=opacity, color='r', label='Ventas')

plt.xlabel('Materiales')
plt.ylabel('Millones')
plt.title('Comparacin de Exportaciones y Ventas por Material')
plt.xticks([i + bar_width / 2 for i in index], materiales)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

materiales = result_df.index
ratios = result_df['ratio_ventas/exports']

fig, ax = plt.subplots(figsize=(10, 6))

bars = plt.bar(materiales, ratios, color='skyblue')

plt.xlabel('Materiales')
plt.ylabel('Ratio Ventas/Exportaciones')
plt.title('Ratio Ventas/Exportaciones por Material')
plt.xticks(rotation=45)
plt.ylim(0, 2) 
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
